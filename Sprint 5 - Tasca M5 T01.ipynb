### Exercici 1

Descarrega el dataset adjunt de dades oficials de la UEFA i selecciona un atribut del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%. 


```python
# Carreguem el dataset de la UEFA

import pandas as pd

pd.read_csv('Lligues europees UEFA.csv', encoding ="ISO-8859-1")
```

```python
# Ordenem les columnes i les dades del dataframe: eliminem ";"

Uefa_df = pd.read_csv('Lligues europees UEFA.csv', encoding ="ISO-8859-1", sep = ';')

pd.set_option('display.max_rows', None)

pd.set_option('display.max_columns', None)

Uefa_df
```

Utilitzem el ___T-STUDENT TEST___

- Comprova si les mitjanes son sensiblement diferents

Plantegem les hipotesis:

- H0: La mitjana de partits empatats = 10
- H1: La mitjana de partits empatats $\not=$ 10

```python
# Seleccionem la variable "D" (partits empatats)

from scipy.stats import ttest_1samp
from statistics import mean


Partits_empatats = Uefa_df['D']

Mitjana_H0 = 10 #(pop_mean)

# t_statistic = (mean - pop_mean)/se (se = standard error)

t_statistic, p =ttest_1samp(Partits_empatats, Mitjana_H0)

print('t_statistic=%.3f, p=%.3f' % (t_statistic, p))

print("t-statistic:", t_statistic)
print("p-valor:", p)

# Comparem el p-valor amb alpha (agafem alpha =0.05 (5%))

if p < 0.05:
    print("Rebutgem hipòtesis nul·la: la mitjana de la mostra de partits empatats és significativament diferent a la mitjana de la mitjana de la hipòtesi nul·la")
else: 
    print("NO podem rebutjar la hipótesi nul·la: la mitjana de la mostra de partits empatats NO és significativament diferent a la mitjana de la mitjana de la hipòtesi nul·la")

    
# Calculem al mitjana de la columna 'D' (partits empatats)
    
print(mean(Uefa_df['D']))  
```

Comentaris:

- Si el valor t-statistic és POSITIU, la mitjana de la mostra és superior a la mitjana poblacional (population mean)
- Si el valor t-statistic és NEGATIU, la mitjana de la mostra és inferior a la mitjana poblacional (population mean)

### Exercici 2

Amb el mateix dataset selecciona dos altres atributs del conjunt de dades. Calcula els p-valors i digues si rebutgen la hipòtesi nul·la agafant un alfa de 5%.


Utilitzem el ___Chi-Squared Test___

- Comprova si dues variables categòriques estan relacionades o son independents

Plantegem les hipotesis:

- H0: Les dues mostres son independents
- H1: Existeix una dependencia entre les mostres 



```python
# Seleccionem les variables "W" (partits guanyats) i GD (diferencia de gols entre gols a favor i gols en contra)

import pandas as pd
from scipy.stats import chi2_contingency

# Generem una taula de contingencia (crosstab) amb les columnes 'W' i 'GD' (taula de freqüències) 

Gols = pd.crosstab(Uefa_df['W'], Uefa_df['GD'])

Gols

# Caluculem

stat, p, dof, expected = chi2_contingency(Gols)
print('stat=%.3f, p=%.4f' % (stat, p))

# Comparem el p-valor amb alpha (agafem alpha =0.05 (5%))

if p > 0.05:
    print('Probablement les variables siguin independentts')
else:
    print('Probablement existeixi dependencia entre les variables')
```

Comentaris:

- Sembla que existeix una relació entre el partits guanyats i la diferencia de gols, a major diferencia de gols en positiu (mes a favor que en contra), més partits guanyats.

```python
# Seleccionem ara les variables "L" (partits perduts) i "D" (partits empatats)

import pandas as pd
from scipy.stats import chi2_contingency

# Generem una taula de contingencia (crosstab) amb les columnes 'L' i 'D' (taula de freqüències) 

Partits = pd.crosstab(Uefa_df['L'], Uefa_df['D'])

Partits

# Caluculem

stat, p, dof, expected = chi2_contingency(Partits)
print('stat=%.3f, p=%.4f' % (stat, p))

# Comparem el p-valor amb alpha (agafem alpha =0.05 (5%))

if p > 0.05:
    print('Probablement les variables siguin independents')
else:
    print('Probablement existeixi dependencia entre les variables')
```

Comentaris:

- En aquest cas és poc problable que hi hagi cap relació entre els partits perduts i empatats.

### Exercici 3

Continua amb el conjunt de dades adjunt i selecciona tres atributs del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%.

Utilitzem el ___Analysis of Variance Test (ANOVA)___

- Comprova si les mitjanes de dues o mes mostres son significativament independents

Plantegem les hipotesis:

- H0: Les mitjanes de les mostres son iguals
- H1: Una o mes de les mitjanes de les mostres no son iguals 


```python
from scipy.stats import f_oneway

# Seleccionem les variables 'W' (partits guanyats), 'D' (partits empatats), 'L' (partits perduts)

mostra_1 = Uefa_df['W']
mostra_2 = Uefa_df['D']
mostra_3 = Uefa_df['L']

# Calculem

stat, p = f_oneway(mostra_1, mostra_2, mostra_3)
print('stat=%.3f, p=%.10f' % (stat, p))

# Comparem el p-valor amb alpha (agafem alpha =0.05 (5%))

if p > 0.05:
    print('Probablement tenen la mateixa distribució')
else:
    print('Probablement tenen diferents distribucions')
```

Comentaris:

- En aquest cas el p-valor és gairabé 0, per tant la hipòtesi nula (H0), pot rebutjar-se, reafirmant la hipòtesi alternativa (H1)
